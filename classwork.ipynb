#1. Most expensive car for each car company make.

#2.Most expensive convertible vehicle for each company.

#3.List all the cars which are automatic.

#4. Premium cars of each car company

# 5. Average price of cars each car company make.

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("car.csv")
```

## Most expensive car for each car company make.

```python
df1=df.groupby("Make")["MSRP"].max()
print(df1)
```

## Most expensive convertible vehicle for each company

```python
df2 = df[df["Vehicle Style"] == "Convertible"].groupby("Make")["MSRP"].max()
print(df2)
```

## List all the cars which are automatic.

```python
df3 = df[df["Transmission Type"] == "AUTOMATIC"]
display(df3)
```

# Premium cars of each car company

```python
df4 = df[df["Engine Fuel Type"].str.contains("premium", na=False)]
df4
```

# Average price of cars each car company make.

```python
df5 = df.groupby("Make")["MSRP"].mean().astype(int)
df5
```
